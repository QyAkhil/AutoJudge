```python
import pandas as pd

df=pd.read_csv("../data/codeforces_data.csv")
df
```

```python
df.columns
```

dropping features not appropriate for the task.

```python
useful_df=df.drop(columns=['contest_id', 'contest_name', 'contest_type','contest_start', 'contest_start_year','interaction_format','editorial',
                           'testset_size', 'official_tests', 'official_tests_complete','aliases','id',
                            'input_mode', 'generated_checker', 'executable', 'generated_tests','note','index','examples'])
```

```python
useful_df.columns
```

Dropping null value.

```python
useful_df.isnull().sum()
```

```python
df_clean = useful_df.dropna(subset=["rating"]).copy()
```

```python
df_clean.value_counts
```

As the dataset does not contain classes it is made using rating provided.

```python
def rating_to_difficulty(r):
    
    if r <= 1200:
        return "Easy"
    elif r <= 1900:
        return "Medium"
    else:
        return "Hard"
df_clean["difficulty"] = df_clean["rating"].apply(rating_to_difficulty)
df_clean["difficulty"].value_counts(normalize=True)
```

combining all text data

```python
df_clean["combined_text"] = (
    df_clean['title'].fillna('') + ' ' +
    df_clean['description'].fillna('') + ' ' +
    df_clean['input_format'].fillna('') + ' ' +
    df_clean['output_format'].fillna('')
)
```

```python
df_clean['combined_text'][34]
```

```python
df_clean['combined_text'] = df_clean['combined_text'].str.replace(r'[_]+', ' ', regex=True)
df_clean['combined_text'] = df_clean['combined_text'].str.replace(r'[^\w\s]',' ', regex=True)
```

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

df_clean['tokens'] = df_clean['combined_text'].apply(word_tokenize)
```

```python
df_clean['tokens']
```

removing stopwords while keeping the ones which might me useful and providing meaning for the problem 

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
keep_words = {
    'if', 'no', 'not', 'nor', 'only', 'exists',
    'more', 'less', 'most', 'least', 'few', 'many',
    'find', 'print', 'given',
    'graph', 'path', 'edge', 'vertex', 'simple', 'constraint'
}

stop_words = stop_words - keep_words
import string
df_clean['tokens_clean'] = df_clean['tokens'].apply(lambda t: [w for w in t if w not in stop_words and w not in string.punctuation])
```

```python
df_clean['tokens_clean'][0]
```

count of algo words might be useful as it topic of question sometime decide difficulty like more topics involved means tougher question.

```python
algo_keywords = {
    'dp', 'greedy', 'binary', 'search', 'dfs', 'bfs',
    'dijkstra', 'segment', 'tree', 'fenwick',
    'bitmask', 'combinatorics', 'modulo',
    'prime', 'graph', 'shortest', 'lca'
}

df_clean['algo_count'] = df_clean['tokens_clean'].apply(lambda t: len(set(t) & algo_keywords))
```

```python
df_clean['rating'].astype(int)
```

we can clearly see as count of algo count increases rating of the problem increases.

```python
df_clean.groupby('algo_count')[['rating']].mean()  
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

df_clean['tokens_lemma'] = df_clean['tokens_clean'].apply(lambda t: [lemmatizer.lemmatize(w) for w in t])
```

```python
df_clean['word_count']=df_clean['tokens'].apply(len)
```

```python
df_clean['word_count']
```

DIfficulty of problem increasing with increase of word count.

```python
df_clean.groupby('difficulty')[['word_count']].mean()
```

```python
df_clean['token_count']=df_clean['tokens_clean'].apply(len)
```

count of complexity terms might be useful as it topic of question sometime decide difficulty  .

```python
complexity_words = {
    'efficient', 'optimize', 'fast', 'logarithmic',
    'linear', 'quadratic', 'time' ,'complexity',
    'constraints', 'limit','limits','bound','maximum', 'minimum', 'time limit',
    'upper bound'
}

df_clean['complexity_terms'] = df_clean['tokens'].apply(
    lambda t: len(set(t) & complexity_words)
)
```

increase in difficulty with increase in complexity terms.

```python
df_clean.groupby('complexity_terms')[['rating']].mean()  
```

Preparing text for applying tfidf 

```python
df_clean['tfidf_text'] = df_clean['tokens_lemma'].apply(lambda x: ' '.join(x))
```

saving all preprocessed data

```python
df_clean.to_pickle('../data/preprocessed_data.pkl')
```
